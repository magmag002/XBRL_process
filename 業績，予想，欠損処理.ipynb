```python
# モジュールのインポート
import pandas as pd
import os
import tqdm
import json
```

```python
col_dic={"number":0,"prediction":0,"pre_term":0}
```

```python
path='new2/df_sum/err/asset_pre'
# xbrl_dfフォルダの中身の名前を取得
folderlist = os.listdir(path)
folderlist=sorted(folderlist)

com_list=[]
for i in folderlist:
    if len(i)==9:
        com_list.append(i)

acvlist=com_list
```

```python
acvlist
```

```python
#acvlist=acvlist[1:]
```

```python
#nanをnoneに
```

```python
for com in tqdm.tqdm(acvlist):
    path="new2/df_sum/err/asset_pre/{}".format(com)
    df = pd.read_csv(path, index_col=0,encoding="cp932")
    for col in df.columns:
        if not col in col_dic:
            li=list(df[col])
            for i in range(0,len(li)):
                if str(li[i])=="nan" or li[i]=="―" or li[i]=="ー":
                    li[i]="None"

            df[col]=li
    
    df.to_csv(path,encoding="cp932")
```

```python
#Dateの作成
```

```python
for com in tqdm.tqdm(acvlist):
    path="new2/df_sum/err/asset_pre/{}".format(com)
    df = pd.read_csv(path, index_col=0,encoding="cp932")

    #indexを開く
    json_open= open('TDNet_full_data/API_index/{}.json'.format(com[:-4]), 'r')
    js = json.load(json_open)
    
    #dateframeのindexをnumberに変更
    df["title"]=list(df.index)
    df.index=df["number"]
    #Dateを追加
    df["date"]=list(df.index)
    
    for i in range(len( js['publiclyList'])-1,-1,-1):
        if int(js['publiclyList'][i]["disclosureNumber"]) in df.index:
            df.loc[int(js['publiclyList'][i]["disclosureNumber"]),"date"]=js['publiclyList'][i]["disclosedDate"]
    #break
    df.to_csv(path,encoding="cp932")
```
